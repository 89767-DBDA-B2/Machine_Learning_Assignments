Que1.-
Predict if a person would buy life insurance based on his age using logistic regression.
About Dataset - Age: Age of a person
bought_insurance : whether person buys Insurance or not (0 - No, 1 - Yes)
predict it for Age : 58,30,20,49,53

```python
import warnings
warnings.filterwarnings(action="ignore")
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("./insurance_data.csv")
df.head(5)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.corr()
```

```python
df.describe()
```

```python
# split the data 
x = df.drop('bought_insurance', axis=1) 

y = df['bought_insurance']
```

```python
# train test split 

from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(
    x,y,train_size=0.8, random_state=123456
)
```

```python
from sklearn.linear_model import LogisticRegression 

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
predictions = model.predict([[58],[30],[20],[49],[53]])
```

```python
predictions
```

```python
plt.scatter(x_test, y_pred)
plt.xlabel('X Test')
plt.ylabel('Predicted Y')
plt.title('Scatter Plot of X_test vs Y_pred')
plt.show()

```
